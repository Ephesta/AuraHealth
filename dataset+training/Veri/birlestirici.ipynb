```python
import pandas as pd
import os

csv_klasoru = "csv" 
birlesik_dosya = "tum_veri.csv" 

veri_listesi = []

for dosya_adi in os.listdir(csv_klasoru):
    if dosya_adi.endswith(".csv"):
        dosya_yolu = os.path.join(csv_klasoru, dosya_adi)
        try:
            df = pd.read_csv(dosya_yolu)
            veri_listesi.append(df)
            print(f"{dosya_adi} yüklendi. Satır sayısı: {len(df)}")
        except Exception as e:
            print(f"HATA - {dosya_adi} okunamadı: {e}")

if veri_listesi:
    df_birlesik = pd.concat(veri_listesi, ignore_index=True)
    df_birlesik.to_csv(birlesik_dosya, index=False, encoding="utf-8-sig")
    print(f"\n Başarılı: {birlesik_dosya} dosyası oluşturuldu ({df_birlesik.shape[0]} satır).")
else:
    print("olmadı")
```
