```python
import os
import json
import pandas as pd

json_klasoru = "json" 
csv_klasoru = "csv"    
os.makedirs(csv_klasoru, exist_ok=True) 

for dosya_adi in os.listdir(json_klasoru):
    if dosya_adi.endswith(".json"):
        json_yolu = os.path.join(json_klasoru, dosya_adi)

        with open(json_yolu, "r", encoding="utf-8") as f:
            veri = json.load(f)

        try:
            daily = veri["daily"]
            df = pd.DataFrame(daily)

            il_adi = dosya_adi.replace(".json", "").replace("_veri", "")
            df["il"] = il_adi

            csv_yolu = os.path.join(csv_klasoru, dosya_adi.replace(".json", ".csv"))
            df.to_csv(csv_yolu, index=False, encoding="utf-8-sig")
            print(f"{dosya_adi} → CSV'ye dönüştürüldü.")

        except Exception as e:
            print(f"Hata: {dosya_adi} işlenemedi. {e}")
```
